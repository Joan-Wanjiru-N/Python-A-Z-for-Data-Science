```python
#calculating recency
sales_user2 =  pd.DataFrame(req_sales_df['CustomerKey'].unique())
sales_user2.columns = ['CustomerKey']

sales_user2
```

## Calculating Recency Score

```python
import matplotlib.pyplot as plt

sse = {}
sales_recency = sales_user2[['Recency']]

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(sales_recency)
    sales_recency["clusters"] = kmeans.labels_
    sse[k] = kmeans.inertia_
    
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of clusters")
plt.show()
```

#### from the above elbow method we select 4 as the number of clusters

```python
#recency score
kmeans = KMeans(n_clusters=4)
kmeans.fit(sales_user2[['Recency']])
sales_user2['RecencyCluster'] = kmeans.predict(sales_user2[['Recency']])

#function for ordering the cluster number
def order_cluster(cluster_field_name, target_field_name,df,ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] =  df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final

sales_user2 = order_cluster('RecencyCluster','Recency',sales_user2,False)

sales_user2.RecencyCluster.describe()
```

## Calculating Frequency Scores

```python
#frequency score
#get order counts for each user and create a dataframe with it
sales_frequency = sales_6m.groupby('CustomerKey').OrderDate.count().reset_index()
sales_frequency.columns = ['CustomerKey','Frequency']

#add this data to our main dataframe
sales_user2 = pd.merge(sales_user2, sales_frequency, on='CustomerKey')
```

```python
#k-means
kmeans = KMeans(n_clusters=4)
kmeans.fit(sales_user2[['Frequency']])
sales_user2['FrequencyCluster'] = kmeans.predict(sales_user2[['Frequency']])

#order the frequency cluster
sales_user2 = order_cluster('FrequencyCluster', 'Frequency',sales_user2,True)

#see details of each cluster
sales_user2.groupby('FrequencyCluster')['Frequency'].describe()
```

## Calculating Revenue Scores

```python
#calculate revenue for each customer
sales_revenue = sales_6m.groupby('CustomerKey').SalesAmount.sum().reset_index()

#merge it with our main dataframe
sales_user2 = pd.merge(sales_user2, sales_revenue, on='CustomerKey')

sales_user2
```

```python
#apply clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(sales_user2[['SalesAmount']])
sales_user2['RevenueCluster'] = kmeans.predict(sales_user2[['SalesAmount']])


#order the cluster numbers
sales_user2 = order_cluster('RevenueCluster', 'SalesAmount',sales_user2,True)

#show details of the dataframe
sales_user2.groupby('RevenueCluster')['SalesAmount'].describe()
```

## Overall Scores

```python
#calculate overall score and use mean() to see details
sales_user2['OverallScore'] = sales_user2['RecencyCluster'] + sales_user2['FrequencyCluster'] + sales_user2['RevenueCluster']
sales_user2.groupby('OverallScore')['Recency','Frequency','SalesAmount'].mean()
```

```python
#create a dataframe with CustomerID and Invoice Date
sales_day_order = sales_6m[['CustomerKey','OrderDate']]
#convert Invoice Datetime to day
sales_day_order['OrderDay'] = sales_6m['OrderDate'].dt.date
sales_day_order = sales_day_order.sort_values(['CustomerKey','OrderDate'])
#drop duplicates
sales_day_order = sales_day_order.drop_duplicates(subset=['CustomerKey','OrderDay'],keep='first')
```

```python
#shifting last 3 purchase dates
sales_day_order['PrevOrderDate'] = sales_day_order.groupby('CustomerKey')['OrderDay'].shift(1)
sales_day_order['T2OrderDate'] = sales_day_order.groupby('CustomerKey')['OrderDay'].shift(2)
sales_day_order['T3OrderDate'] = sales_day_order.groupby('CustomerKey')['OrderDay'].shift(3)
```

```python
sales_day_order.info()
```

```python
sales_day_order['DayDiff'] = (sales_day_order['OrderDay'] - sales_day_order['PrevOrderDate']).dt.days
sales_day_order['DayDiff2'] = (sales_day_order['OrderDay'] - sales_day_order['T2OrderDate']).dt.days
sales_day_order['DayDiff3'] = (sales_day_order['OrderDay'] - sales_day_order['T3OrderDate']).dt.days
```

```python
sales_day_order
```

```python
sales_day_diff = sales_day_order.groupby('CustomerKey').agg({'DayDiff': ['mean','std']}).reset_index()
sales_day_diff.columns = ['CustomerKey', 'DayDiffMean','DayDiffStd']
```

```python
#keep customers who have >3 purchases
sales_day_order_last = sales_day_order.drop_duplicates(subset=['CustomerKey'],keep='last')
```

```python
'''0–20: Customers that will purchase in 0–20 days — Class name: 2
21–49: Customers that will purchase in 21–49 days — Class name: 1
≥ 50: Customers that will purchase in more than 50 days — Class name: 0
sales_class['NextPurchaseDayRange'] = 2'''

sales_class.loc[sales_class.NextPurchaseDay>50,'NextPurchaseDayRange'] = 1
sales_class.loc[sales_class.NextPurchaseDay>100,'NextPurchaseDayRange'] = 0

sales_class
```

```python
#sales_class = sales_class.drop('NextPurchaseDay',axis=1)
sales_class.dropna(inplace=True)
sales_class.isna().any()
```

```python
#train & test split
sales_class = sales_class.drop('NextPurchaseDay',axis=1)
X, y = sales_class.drop('NextPurchaseDayRange',axis=1), sales_class.NextPurchaseDayRange
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)

#create an array of models
models = []
models.append(("LR",LogisticRegression()))
models.append(("NB",GaussianNB()))
models.append(("RF",RandomForestClassifier()))
models.append(("SVC",SVC()))
models.append(("Dtree",DecisionTreeClassifier()))
#models.append(("XGB",xgb.XGBClassifier()))
models.append(("KNN",KNeighborsClassifier()))

#measure the accuracy 
for name,model in models:
    kfold = KFold(n_splits=2, random_state=22)
    cv_result = cross_val_score(model,X_train,y_train, cv = kfold,scoring = "accuracy")
    print(name, cv_result)
```
